1. in functional api model there is possible that we can have multiple inputs of different data types like tabular data , text data , and img data and we have to predict one thing/one output .
2. this is also possible that we have one or two inputs and multiple outputs .

so in these above cases we can't use sequential models we use functinal api models for that

```python
from keras.models import Model
```

```python
x = Input(shape=(3,))
```

```python
model = Model(inputs = x, outputs=[output1, output2] )
```

```python
from keras.layers import *
```

```python
hidden1 = Dense(3, activation='relu')(x)
hidden2 = Dense(3, activation='relu')(hidden1)

output1 = Dense(1, activation='relu')(hidden2)
output2 = Dense(10, activation='sigmoid')(hidden2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True)
```
