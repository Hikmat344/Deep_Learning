```python
# Replace 'your_file.txt' with the name of your uploaded file
file_name = 'LSTM DATA.txt'  # Automatically get the uploaded file name

# Reading the file content
with open(file_name, 'r') as file:
    text = file.read()

```

```python
import pandas as pd
import numpy as np
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```


1.    Now I will split our entire document into sentences

2.   convert sentences into nubers ( numeric reprentation of sentences )



```python
input_sequences = []
#1
for sentences in text.split("\n"):
  # print(sentences)
#2
  tokanized_sentence = tokenizer.texts_to_sequences([sentences])[0] # here [0] because it is made up a double list so we print the 0th list
  for sent in range(1,len(tokanized_sentence)):
    input_sequences.append(tokanized_sentence[:sent+1])

```

we can see we changed it into input sequences, and this is a joint 2d array. now it is canged into input data . stil we have to change it into input|output format.

# Issue:
issue is that as in supervised  the input dim/size is same but here the size is different for different sentences so now we will add 0 pading to resolve this issue

```python
input_sequences
```

now to add 0 padding , we have to find the max len of the input seq

```python
max_len = max([len(seq) for seq in input_sequences])
max_len
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_inputSequences = pad_sequences(input_sequences, maxlen=max_len, padding="pre") # padding pre , because we want to add zeros before the sentence words
```

now sized has been equaled with each row

now we will extract input and outputs

```python
x = padded_inputSequences[:,:-1]
```

```python
y = padded_inputSequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

now to do onehot encoding on y/output

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
y = to_categorical(y, num_classes=len(tokenizer.word_index)+1) # here +l because tokenizer index starts from 1 whilw onehot encoding starts from 0
```

```python
y.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=max_len-1))
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+1, activation="softmax"))
```

```python
model.build(input_shape=(None, max_len-1))
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train,epochs=50,batch_size=100,validation_data=(X_val, y_val))  # Provide validation data here
```

```python
model.fit(x, y, epochs=100)
```

# Do prediction

```python
text = 'my'

#1. tokanize
tokanized_sentence = tokenizer.texts_to_sequences([text])[0]

#2. add padding
padded_sentence = pad_sequences([tokanized_sentence], maxlen=10, padding="pre")

#3. prediction
position = np.argmax(model.predict(padded_sentence))
```

```python
for word,index in tokenizer.word_index.items():
  if index == position:
    print(word)
```

```python
import time
text = "my"

for i in range(5):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=10, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
